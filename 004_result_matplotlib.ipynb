```python
import pandas as pd
data = pd.read_excel('창폐업4사분면.xlsx')
data
```

```python
서울중구 = data[data['구이름']=='서울 중구']
서울노원구 = data[data['구이름']=='서울 노원구']
대구중구 = data[data['구이름']=='대구 중구']
대구수성구 = data[data['구이름']=='대구 수성구']
```

```python
plt.style.use('fivethirtyeight')
```

```python
import matplotlib.pyplot as plt
print(plt.style.available) #가진놈들
```

```python
from matplotlib import font_manager,rc
font_path = "c:/Windows/Fonts/LG_Smart_UI-Regular.TTF"
font_name = font_manager.FontProperties(fname=font_path).get_name() #한글 출력을 위해 셋팅
rc('font',family=font_name)
plt.xlim(-3,4)
plt.ylim(-1,0.2)
plt.scatter(서울중구['창폐업증감'],서울중구['카드매출증감율'])
plt.scatter(서울노원구['창폐업증감'],서울노원구['카드매출증감율'])
plt.scatter(대구중구['창폐업증감'],대구중구['카드매출증감율'])
plt.scatter(대구수성구['창폐업증감'],대구수성구['카드매출증감율'])
plt.legend(['서울 중구','서울 노원구','대구 중구','대구 수성구'],loc=4,fontsize='medium')
```
